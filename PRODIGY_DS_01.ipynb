```python
# Drop irrelevant columns and clean up
data = data.drop(columns=["Indicator Name", "Indicator Code", "Unnamed: 69"], errors="ignore")

# Convert wide to long format
df_melted = data.melt(id_vars=["Country Name", "Country Code"], var_name="Year", value_name="Population")

# Convert Year to numeric
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
df_melted = df_melted.dropna(subset=['Year'])

# Example: plot population of India over time
india_df = df_melted[df_melted["Country Name"] == "India"]

fig = px.bar(india_df, x="Year", y="Population", title="India Population Over Time")
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already loaded
# df = pd.read_excel("your_file.xls", sheet_name="Data", engine="xlrd")

# ---------- 1. Bar Chart: Population of a Specific Country Over Time ----------
country_name = "India"
all_years = [str(year) for year in range(1960, 2024)]
decade_years = [str(year) for year in range(1960, 2024, 10)]  # every 10 years

country_data = data[data['Country Name'] == country_name]
if not country_data.empty:
    population = country_data[all_years].values.flatten()

    plt.figure(figsize=(12, 6))
    plt.bar(all_years, population, color='cornflowerblue')
    plt.xticks(ticks=[i for i, y in enumerate(all_years) if y in decade_years],
               labels=[y for y in all_years if y in decade_years],
               rotation=45)
    plt.title(f'Population of {country_name} (1960–2023)')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.tight_layout()
    plt.show()
else:
    print(f"Country '{country_name}' not found in the dataset.")

```

```python
import pandas as pd
import plotly.graph_objs as go

# Define year range for plotting (every 10 years)
years = [str(year) for year in range(1960, 2024, 10)]

# Create the figure
fig = go.Figure()

# Loop through each country and add a trace (line)
for _, row in data.iterrows():
    country = row['Country Name']
    population = row[years].values

    # Add a line per country
    fig.add_trace(go.Scatter(
        x=years,
        y=population,
        mode='lines+markers',
        name=country,
        hovertemplate=f"<b>{country}</b><br>Year: %{x}<br>Population: %{y}<extra></extra>",
        line=dict(width=1)
    ))

# Customize layout
fig.update_layout(
    title="Population Growth of All Countries (Every 10 Years)",
    xaxis_title="Year",
    yaxis_title="Population",
    width=800,
    height=800,
    showlegend=False  # Set True to show all countries in the legend
)

# Show the interactive plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV properly
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

# If "Indicator Name" column exists, filter, else skip
if "Indicator Name" in df.columns:
    pop_df = df[df['Indicator Name'] == 'Population, total'].copy()
else:
    pop_df = df.copy()

# Define decades
decades = {
    '1960s': [str(year) for year in range(1960, 1970)],
    '1970s': [str(year) for year in range(1970, 1980)],
    '1980s': [str(year) for year in range(1980, 1990)],
    '1990s': [str(year) for year in range(1990, 2000)],
    '2000s': [str(year) for year in range(2000, 2010)],
    '2010s': [str(year) for year in range(2010, 2020)],
    '2020s': [str(year) for year in range(2020, 2024)],
}

# Aggregate population by decade
for decade, years in decades.items():
    available_years = [y for y in years if y in pop_df.columns]
    pop_df[decade] = pop_df[available_years].sum(axis=1)

# Example 1: Bar chart for one country across decades
country_name = 'India'
country_data = pop_df[pop_df['Country Name'] == country_name][list(decades.keys())].T

plt.figure(figsize=(10,6))
plt.bar(country_data.index, country_data.values.flatten())
plt.title(f'Population of {country_name} Across Decades')
plt.xlabel('Decade')
plt.ylabel('Population (Sum)')
plt.show()

# Example 2: Histogram of population distribution for all countries in the 2010s
plt.figure(figsize=(10,6))
plt.hist(pop_df['2010s'].dropna(), bins=20, edgecolor='black')
plt.title('Population Distribution Across Countries in the 2010s')
plt.xlabel('Population (Sum)')
plt.ylabel('Number of Countries')
plt.show()
```
